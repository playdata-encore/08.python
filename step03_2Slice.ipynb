### Slice

첨부된 이미지 보고 소스 이해하기

1. 시퀀스객체[시작인텍스:끝인덱스]
- 시퀀스객체[:끝인덱스] : 끝인덱스-1까지 가져옴
- 시퀀스객체[시작인덱스:] : 해당 index 부터 마지막 요소까지 가져옴
- 시퀀스객체[:] : 리스트 전체를 가져옴
- 시퀀스객체[시작:끝:증가폭]
- 시퀀스객체[:: 증가폭] : 리스트 전체에서 증가폭씩 증가시키면서 요소를 가져옴
- 시퀀스객체[::] : 시스트 전체를 가져옴

```python
import requests
import random
from bs4 import BeautifulSoup
import webbrowser

results = []
req = requests.get('https://www.google.com/search?tbm=isch&q=' + input('검색어를 입력하세요 : '))
html = req.text
soup = BeautifulSoup(html, 'html.parser')

for i in str(soup).split(' '):
    if i.startswith('src'):
        results.append(i.split('"')[1])

print(results[int(random.random()*len(results))])
webbrowser.open(results[int(random.random()*len(results))])
```
